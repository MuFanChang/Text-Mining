```python
import csv
import os
import sys
import operator

#***(1) 讀入資料後，使用tokens=sentence.split() : 使用此方法將sentence斷開，取得每一行的tokens。用於文本、停止詞、標點符號。***
fp = open("TMBD_news_files_assignment1.txt","r")
news = fp.read()
tokens = news.split()

stopword = open("stopword_chinese.txt","r")
stopwords = stopword.read()
sws = stopwords.split()

punctuation = open("punctuation.txt","r")
punc = punctuation.read()
pc = punc.split()

#***(2) 定義function <-return [w for w in tokens if w not in sws] 用以刪除停止詞和標點符號。***
def removeStopwords(tokens,sws):
    return [w for w in tokens if w not in sws]
rm1 = removeStopwords(tokens=tokens,sws=sws)
rm2 = removeStopwords(tokens=rm1,sws=pc)

#***(3) 將清理過後的文本裝入dict，跑for迴圈計數。***
word_counts = {}
for word in rm2:
    if word.upper() in word_counts.keys():
        word_counts[word.upper()] = word_counts[word.upper()] + 1
    else:
        word_counts[word.upper()] = 1

#***(4) 依據dict中value排序（大到小）。***
sorted_wc = sorted(word_counts.items(), key=operator.itemgetter(1),reverse=True)
sorted_wc
```

```python
f = open('dictionary.txt', 'w', encoding= 'UTF-8')  
f.write('\n'.join(map(lambda x: str(x),sorted_wc)))
f.close()
```
